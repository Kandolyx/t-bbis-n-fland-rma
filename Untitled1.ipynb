```python
# 📦 Gerekli kütüphaneleri yükleyin (sadece 1 kez çalıştırmak yeterli)
!pip install xgboost lightgbm transformers sentence-transformers umap-learn plotly scikit-learn wordcloud imbalanced-learn torch --upgrade
!pip install timm

# 📁 Google Drive bağlantısı
from google.colab import drive
drive.mount('/content/drive')

# 🔧 Gerekli kütüphaneler
import numpy as np
import pandas as pd
import torch
import os
import time
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.linear_model import LogisticRegression
from sklearn.svm import LinearSVC
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from lightgbm import LGBMClassifier
from sklearn.metrics import classification_report, accuracy_score, f1_score
from tqdm import tqdm
from transformers import AutoTokenizer, AutoModel
from scipy.sparse import hstack
from imblearn.over_sampling import SMOTE
import matplotlib.pyplot as plt
import re
import seaborn as sns
from wordcloud import WordCloud
def clean_text(text):
    text = text.lower()                            # Küçük harfe çevir
    text = re.sub(r"http\S+", "", text)            # URL temizle
    text = re.sub(r"\d+", " ", text)               # Sayıları kaldır
    text = re.sub(r"[^\w\s]", " ", text)           # Noktalama kaldır
    text = re.sub(r"\s+", " ", text).strip()       # Ekstra boşluklar
    return text

def load_and_clean_data(path, is_train=True):
    df = pd.read_csv(path, sep='\t', header=None)
    df.columns = ['label', 'text'] if is_train else ['text']
    df['text'] = df['text'].astype(str).apply(clean_text)
    return df


def analyze_data(df):
    # Önce label sütunundaki eşsiz değerleri yazdır (sorunu anlamak için)
    print("Label sütunundaki unique değerler:", df['label'].unique())

    label_map = {
        "1": "Infectious diseases",
        "2": "Congenital anomalies",
        "3": "Other diagnoses",
        "4": "Symptoms",
        "5": "Neoplasms"
    }

    df['label_name'] = df['label'].astype(str).map(label_map)
    label_counts = df['label_name'].value_counts()

    print("Label counts:\n", label_counts)  # kontrol için

    plt.figure(figsize=(8, 5))
    sns.barplot(x=label_counts.index, y=label_counts.values)
    plt.xticks(rotation=45)
    plt.title("Sınıf Dağılımı (Orijinal Veri)")
    plt.xlabel("Label")
    plt.ylabel("Adet")
    plt.tight_layout()
    plt.show()

    for label in label_counts.index:
        subset = df[df['label_name'] == label]
        text = " ".join(subset['text'].tolist())
        wordcloud = WordCloud(width=800, height=400, background_color='white').generate(text)
        plt.figure(figsize=(10, 5))
        plt.imshow(wordcloud, interpolation='bilinear')
        plt.axis("off")
        plt.title(f"WordCloud - {label} (Orijinal Veri)")
        plt.show()

def plot_class_distribution(y_train, y_train_smote, le):
    # SMOTE öncesi ve sonrası sınıf dağılımlarını karşılaştır
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 6))

    # SMOTE öncesi
    unique, counts = np.unique(y_train, return_counts=True)
    sns.barplot(x=le.inverse_transform(unique), y=counts, ax=ax1)
    ax1.set_title("Sınıf Dağılımı (SMOTE Öncesi)")
    ax1.set_xlabel("Sınıf")
    ax1.set_ylabel("Örnek Sayısı")
    ax1.tick_params(axis='x', rotation=45)

    # SMOTE sonrası
    unique_smote, counts_smote = np.unique(y_train_smote, return_counts=True)
    sns.barplot(x=le.inverse_transform(unique_smote), y=counts_smote, ax=ax2)
    ax2.set_title("Sınıf Dağılımı (SMOTE Sonrası)")
    ax2.set_xlabel("Sınıf")
    ax2.set_ylabel("Örnek Sayısı")
    ax2.tick_params(axis='x', rotation=45)

    plt.tight_layout()
    plt.show()

def get_models(xgb_params, num_classes):
    lgb_params = {
        "boosting_type": "gbdt",
        "objective": "multiclass",
        "num_class": num_classes,
        "n_estimators": 100,
        "random_state": 42
    }
    return {
        "XGBoost": XGBClassifier(**xgb_params, use_label_encoder=False, eval_metric='mlogloss'),
        "Linear SVM": LinearSVC(max_iter=1000),
        "LogisticRegression": LogisticRegression(max_iter=500),
        "RandomForest": RandomForestClassifier(n_estimators=100),
        "LightGBM": LGBMClassifier(**lgb_params)
    }

def get_enhanced_embeddings(texts, model_name="prajjwal1/bert-tiny", batch_size=32, cache_file="embeddings.npy"):
    if os.path.exists(cache_file):
        print(f"📁 Cache'den embedding'ler yükleniyor: {cache_file}")
        return np.load(cache_file)

    tokenizer = AutoTokenizer.from_pretrained(model_name)
    model = AutoModel.from_pretrained(model_name)
    model.eval()
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    model.to(device)

    all_embeddings = []
    for i in tqdm(range(0, len(texts), batch_size), desc="BERT Batch İşleniyor"):
        batch = texts[i:i+batch_size]
        encoded = tokenizer(batch, padding=True, truncation=True, return_tensors="pt", max_length=128)
        encoded = {key: val.to(device) for key, val in encoded.items()}

        with torch.no_grad():
            outputs = model(**encoded)
            embeddings = outputs.last_hidden_state[:, 0, :].cpu().numpy()
            all_embeddings.append(embeddings)

    final_embeddings = np.vstack(all_embeddings)
    np.save(cache_file, final_embeddings)
    print(f"💾 Embedding'ler kaydedildi: {cache_file}")
    return final_embeddings

def extract_text_features(texts):
    features = []
    for text in texts:
        length = len(text)
        word_count = len(text.split())
        upper_ratio = sum(1 for c in text if c.isupper()) / length if length > 0 else 0
        digit_ratio = sum(1 for c in text if c.isdigit()) / length if length > 0 else 0
        features.append([length, word_count, upper_ratio, digit_ratio])
    return np.array(features)

def get_enhanced_features(texts, use_tfidf=True, use_embeddings=True, embedding_cache="embeddings.npy"):
    features = []

    if use_embeddings:
        embeddings = get_enhanced_embeddings(texts, cache_file=embedding_cache)
        features.append(embeddings)

    if use_tfidf:
        tfidf = TfidfVectorizer(max_features=1000, ngram_range=(1, 1))
        X_tfidf = tfidf.fit_transform(texts)
        features.append(X_tfidf)

    extra_features = extract_text_features(texts)
    if use_tfidf:
        return hstack([*features, extra_features])
    else:
        return np.hstack([*features, extra_features])

def train_and_evaluate(models, X_train, y_train, X_test, y_test, label_encoder, smote_status):
    results = []

    for name, model in models.items():
        print(f"\n🎯 {name} eğitiliyor... (SMOTE: {smote_status})")
        start = time.time()
        model.fit(X_train, y_train)
        end = time.time()
        preds = model.predict(X_test)

        acc = accuracy_score(y_test, preds)
        f1 = f1_score(y_test, preds, average='weighted')
        duration = end - start

        results.append({
            "Model": name,
            "Accuracy": acc,
            "F1 Score": f1,
            "Training Time (s)": duration,
            "SMOTE": smote_status
        })

        print(f"⏱️ Eğitim Süresi: {duration:.2f} saniye")
        print(classification_report(y_test, preds, target_names=label_encoder.classes_.astype(str)))

    return pd.DataFrame(results)

def plot_comparison_results(all_results_df):
    # 📈 Model Karşılaştırma Grafikleri
    plt.figure(figsize=(14, 6))
    sns.barplot(data=all_results_df, x="Model", y="Accuracy", hue="SMOTE")
    plt.title("Model Karşılaştırması - Doğruluk (SMOTE Öncesi ve Sonrası)")
    plt.legend(title="SMOTE Uygulandı mı?")
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

    plt.figure(figsize=(14, 6))
    sns.barplot(data=all_results_df, x="Model", y="F1 Score", hue="SMOTE")
    plt.title("Model Karşılaştırması - F1 Skor (SMOTE Öncesi ve Sonrası)")
    plt.legend(title="SMOTE Uygulandı mı?")
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

    plt.figure(figsize=(14, 6))
    sns.barplot(data=all_results_df, x="Model", y="Training Time (s)", hue="SMOTE")
    plt.title("Model Eğitim Süreleri (SMOTE Öncesi ve Sonrası)")
    plt.legend(title="SMOTE Uygulandı mı?")
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

def main():
    # 📌 Parametreler
    if torch.cuda.is_available():
        xgb_params = {'tree_method': 'gpu_hist', 'predictor': 'gpu_predictor', 'gpu_id': 0}
        print("🚀 GPU kullanılacak")
    else:
        xgb_params = {'tree_method': 'hist', 'predictor': 'cpu_predictor'}
        print("🖥️ CPU kullanılacak")

    # 📂 Veri Yükleme ve Temizleme
    df = load_and_clean_data('/content/drive/MyDrive/medical/train.dat', is_train=True)
    df['label'] = df['label'].astype(str)

    # 📊 Sınıf dağılımı ve WordCloud
    analyze_data(df)

    # 🎯 Label encode
    le = LabelEncoder()
    df['label_encoded'] = le.fit_transform(df['label'])

    texts = df['text'].tolist()
    labels = df['label_encoded'].tolist()
    num_classes = len(le.classes_)

    # 🧪 3 özellik seti: TF-IDF, BERT, BERT + TF-IDF
    feature_configs = [
        {"name": "TF-IDF", "use_tfidf": True, "use_embeddings": False, "cache": "tfidf_dummy.npy"},
        {"name": "BERT", "use_tfidf": False, "use_embeddings": True, "cache": "bert_only.npy"},
        {"name": "BERT + TF-IDF", "use_tfidf": True, "use_embeddings": True, "cache": "bert_tfidf.npy"}
    ]

    all_results = []

    for config in feature_configs:
        print(f"\n⚙️ Özellik Seti: {config['name']}")
        X = get_enhanced_features(
            texts,
            use_tfidf=config["use_tfidf"],
            use_embeddings=config["use_embeddings"],
            embedding_cache=config["cache"]
        )
        y = np.array(labels)

        X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)

        # SMOTE öncesi sınıf dağılımı
        plot_class_distribution(y_train, y_train, le)

        # 🧠 SMOTE öncesi modeller
        models = get_models(xgb_params, num_classes=num_classes)
        results_df_no_smote = train_and_evaluate(models, X_train, y_train, X_test, y_test, le, "Hayır")
        results_df_no_smote["Feature Set"] = config["name"]
        all_results.append(results_df_no_smote)

        # 🔁 SMOTE uygulama
        print("🔁 SMOTE uygulanıyor...")
        smote = SMOTE(random_state=42)
        X_train_smote, y_train_smote = smote.fit_resample(X_train, y_train)

        # SMOTE sonrası sınıf dağılımı
        plot_class_distribution(y_train, y_train_smote, le)

        # 🧠 SMOTE sonrası modeller
        models = get_models(xgb_params, num_classes=num_classes)
        results_df_smote = train_and_evaluate(models, X_train_smote, y_train_smote, X_test, y_test, le, "Evet")
        results_df_smote["Feature Set"] = config["name"]
        all_results.append(results_df_smote)

    # Tüm sonuçları birleştir
    all_results_df = pd.concat(all_results, ignore_index=True)

    # 📊 Sonuçları çizdir
    plot_comparison_results(all_results_df)

    print("\n📊 Karşılaştırma Tablosu:")
    print(all_results_df)


if __name__ == "__main__":
    main()
```
